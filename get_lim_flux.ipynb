```python
import astropy.units as u
from astropy.constants import c
# AB magnitude to flux in erg/s/cm^2/Hz
def ABmag2flux(mag):
    return 10**(-(mag+48.6)/2.5)*u.erg/u.s/u.cm**2/u.Hz

fl_hz_19 = ABmag2flux(19)
fl_hz_22 = ABmag2flux(22)
fl_hz_23 = ABmag2flux(23.2)
```

```python
fl_a_19 = (fl_hz_19*c/(6000*u.AA)**2).to(u.erg/u.s/u.cm**2/u.AA)
fl_a_22 = (fl_hz_22*c/(6000*u.AA)**2).to(u.erg/u.s/u.cm**2/u.AA)
fl_a_23 = (fl_hz_23*c/(6000*u.AA)**2).to(u.erg/u.s/u.cm**2/u.AA)
```
